what is Count Plot

count plot are used to when we have a categorical data , variables , and you want to visulize the distribution of these categories.

```python
sns.countplot(x='class',data=df)
```

```python
sns.countplot(x='source_city',data=df)
```

2. Piechart (use for a categorial data )

a pie chart represent the part of a whole entity.

```python
df['source_city'].value_counts().plot(kind='pie',autopct='%.2f')
```

Donut plots (used for categorical data )

donut plot are used to show the preporation of each category in relation to the whole.

```python
plt.pie(df['source_city'].value_counts(),labels=df['source_city'].unique(),autopct='%.2f',wedgeprops=dict(width=0.7))
```

Histogram (used for numerical data) (continuous variables:)

implement boxplot ,scatter plot , line plot with suitable data.

```python

plt.figure()
df.boxplot(column="price", by="airline")
plt.title("Price Distribution by Airline")
plt.suptitle("")
plt.xlabel("Airline")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure()
plt.scatter(df["duration"], df["price"])
plt.xlabel("Duration (Hours)")
plt.ylabel("Price")
plt.title("Duration vs Price")
plt.show()
```

```python
avg_price = df.groupby("days_left")["price"].mean()

plt.figure()
plt.plot(avg_price.index, avg_price.values)
plt.xlabel("Days Left")
plt.ylabel("Average Price")
plt.title("Average Price vs Days Left")
plt.show()
```
